```python
filename = '/data/proposals/info.txt'
filelist = []
for line in open(filename):
    filelist.append(line)
```

```python
print filelist[6]
```

```python
file_dict = dict()
for f in filelist:
    f = f.strip()
    fields = f.split(',')
    if int(fields[3]) == 1:
        # See if field[1] exists in dict
        key = fields[1]
        info = fields[0] + ',' + fields[4] + ',' + fields[5] + ',' + fields[6] + ',' + fields[7]
        if key not in file_dict:
            file_dict[key] = []
        file_dict[key].append(info)
```

```python
print file_dict['/opt/data/leftImg8bit/val/frankfurt/frankfurt_000000_000294_leftImg8bit.png']
```

```python
for k,v in file_dict.iteritems():
    fields = k.strip().split('/')
    filename = fields[-1].replace('leftImg8bit','results')
    print filename
```

```python
filename = '/opt/data/leftImg8bit/val/frankfurt/frankfurt_000000_000576_leftImg8bit.png'
value = file_dict[filename]
img = cv2.imread(filename.replace('png','inst.png'), cv2.IMREAD_GRAYSCALE)
for v in value:
    fields = v.split(',')
    x = int(fields[1])
    y = int(fields[2])
    w = int(fields[3])
    h = int(fields[4])
    cv2.rectangle(img,(x,y),(x+w,y+h),(10),2)
```

```python
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(20,10)
f1 = plt.imshow(img)
print img.shape
plt.show()
```
